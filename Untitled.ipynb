```python
import requests, itertools, pandas as pd, time

API_KEY = "6e16b11542b37f683e1a381a58b1a8172792c11b"
YEAR = "2023"
BASE = f"https://api.census.gov/data/{YEAR}/acs/acs5"
params_geo = "for=tract:*&in=state:36+county:067"

# 1. Get all variables
meta_url = f"{BASE}/variables.json"
meta = requests.get(meta_url).json()["variables"]

# 2. Only use variables available at tract-level
tract_vars = [
    var for var, info in meta.items()
    if info.get("predicateType") in ["int", "float"] and info.get("group") and not var.startswith("GEO")
]

print(f"{len(tract_vars)} tract-level variables identified.")

# 3. Chunk into groups of 40 (to be safe)
def chunks(lst, n):
    for i in range(0, len(lst), n):
        yield lst[i:i + n]

var_chunks = list(chunks(tract_vars, 40))

dfs = []
for i, chunk in enumerate(var_chunks, start=1):
    vars_param = ",".join(chunk)
    url = f"{BASE}?get={vars_param}&{params_geo}&key={API_KEY}"
    resp = requests.get(url)
    retry = 0
    while resp.status_code == 429 and retry < 5:
        time.sleep(2)
        resp = requests.get(url)
        retry += 1
    if resp.status_code != 200:
        print(f"[{i}] Skipping (HTTP {resp.status_code})")
        continue
    try:
        data = resp.json()
        hdr, *rows = data
        if not rows or any(len(row) != len(hdr) for row in rows):
            print(f"[{i}] Skipping malformed chunk")
            continue
        df = pd.DataFrame(rows, columns=hdr)
        dfs.append(df)
        print(f"Fetched chunk {i}/{len(var_chunks)}")
    except Exception as e:
        print(f"[{i}] Error: {e}")
        continue

if not dfs:
    raise RuntimeError("No valid dataframes retrieved.")

# Merge
full = dfs[0]
for df in dfs[1:]:
    full = full.merge(df, on=["state", "county", "tract"])

full.to_csv("onondaga_tract_acs2023_allvars.csv", index=False)
print("Done!", full.shape)
```

```python
import requests
import pandas as pd

# Define the API endpoints
data_url = "https://api.census.gov/data/2022/acs/acs5?get=group(B17001)&ucgid=pseudo(0500000US36067$1400000)"
metadata_url = "https://api.census.gov/data/2022/acs/acs5/groups/B17001.json"

# Fetch the data
response = requests.get(data_url)
data = response.json()

# Fetch the metadata
metadata_response = requests.get(metadata_url)
metadata = metadata_response.json()

# Extract headers and data
headers = data[0]
data_rows = data[1:]

# Create a DataFrame for the data
df = pd.DataFrame(data_rows, columns=headers)

# Create a DataFrame for the metadata
variables = metadata['variables']
metadata_list = []
for key, value in variables.items():
    metadata_list.append({
        'name': key,
        'label': value.get('label', ''),
        'concept': value.get('concept', ''),
        'predicateType': value.get('predicateType', '')
    })

metadata_df = pd.DataFrame(metadata_list)

# Create a data dictionary
data_dictionary = {}
for header in headers:
    meta_info = metadata_df[metadata_df['name'] == header]
    if not meta_info.empty:
        data_dictionary[header] = {
            'label': meta_info['label'].values[0],
            'concept': meta_info['concept'].values[0],
            'predicateType': meta_info['predicateType'].values[0]
        }
    else:
        data_dictionary[header] = {
            'label': 'N/A',
            'concept': 'N/A',
            'predicateType': 'N/A'
        }

# Print the data dictionary
for key, value in data_dictionary.items():
    print(f"{key}:")
    for sub_key, sub_value in value.items():
        print(f"  {sub_key}: {sub_value}")

```

```python
import requests
import pandas as pd

# Define the API endpoints
poverty_url = "https://api.census.gov/data/2022/acs/acs5?get=group(B17001)&ucgid=pseudo(0500000US36067$1400000)"
population_url = "https://api.census.gov/data/2022/acs/acs5?get=group(B01003)&ucgid=pseudo(0500000US36067$1400000)"

# Fetch the data for poverty status
poverty_response = requests.get(poverty_url)
poverty_data = poverty_response.json()

# Fetch the data for total population
population_response = requests.get(population_url)
population_data = population_response.json()

# Extract headers and data
poverty_headers = poverty_data[0]
poverty_rows = poverty_data[1:]
population_headers = population_data[0]
population_rows = population_data[1:]

# Create DataFrames for the data
df_poverty = pd.DataFrame(poverty_rows, columns=poverty_headers)
df_population = pd.DataFrame(population_rows, columns=population_headers)

# Extract relevant columns for poverty metrics
poverty_columns = ['NAME', 'B17001_001E', 'B17001_002E']
df_poverty = df_poverty[poverty_columns]

# Extract relevant columns for total population metrics
population_columns = ['NAME', 'B01003_001E']
df_population = df_population[population_columns]

# Convert relevant columns to numeric
for col in poverty_columns[1:]:
    df_poverty[col] = pd.to_numeric(df_poverty[col], errors='coerce')
df_population['B01003_001E'] = pd.to_numeric(df_population['B01003_001E'], errors='coerce')

# Merge the two datasets on 'NAME'
df_combined = pd.merge(df_poverty, df_population, on='NAME', suffixes=('_poverty', '_population'))

# Calculate the poverty rate
df_combined['Poverty Rate (%)'] = (df_combined['B17001_002E'] / df_combined['B17001_001E']) * 100

# Select only the necessary columns for the final output
df_final = df_combined[['NAME', 'B01003_001E', 'B17001_002E', 'Poverty Rate (%)']]
df_final.columns = ['NAME', 'Total Population', 'Below Poverty', 'Poverty Rate (%)']

# Display the final DataFrame
print(df_final.head())

# Save the final DataFrame to a CSV file
```
